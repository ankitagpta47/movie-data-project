```python
import pandas as pd
df= pd.read_csv('final-project/movies_metadata.csv',low_memory=False)
df.info()

df.head()
```

```python
df.columns
```

```python
--DATA CLEANING
```

```python
# columns we want to keep (use only those that actually exist in df)
useful = [
    "title", "original_title", "release_date",
    "budget", "revenue", "genres", "popularity",
    "runtime", "vote_average", "vote_count",
    "production_companies", "production_countries",
    "imdb_id"  
]


keep = [c for c in useful if c in df.columns]
df_keep = df.loc[:, keep].copy()


import pandas as pd
df_keep["release_date"] = pd.to_datetime(df_keep["release_date"], errors="coerce")
df_keep["release_year"] = df_keep["release_date"].dt.year

df_keep.info()
df_keep.head()
```

```python
df.columns
```

--convert numeric columns

```python
num_cols = ["budget", "revenue", "popularity", "runtime", "vote_average", "vote_count"]

for col in num_cols:
    df_keep[col] = pd.to_numeric(df_keep[col], errors='coerce')
    
    
num_cols
```

```python
--Handle missing values
```

```python
df_keep.isnull().sum()
```

```python

df_keep = df_keep.dropna(subset=['title', 'release_year'])


df_keep['runtime'].fillna(df_keep['runtime'].median(), inplace=True)


df_keep
```

```python
--Remove duplicates
```

```python
df_keep.drop_duplicates(subset=['title', 'release_year'], inplace=True)

df_keep
```

```python
-- Normalize titles
```

```python
import re

def normalize_title(title):
    if pd.isna(title):
        return None
    return re.sub(r'[^a-z0-9 ]', '', title.lower().strip())

df_keep["title_norm"] = df_keep["title"].apply(normalize_title)
```

```python
df_keep.info()
df_keep.head()
```

```python
import pandas as pd
import numpy as np
import re



df2 = pd.read_csv('final-project/box_office.csv')
df2.head()
```

-- Rename columns for easier access

```python

df2.rename(columns={
    "Release Group": "title",
    "Year": "release_year",
    "$Worldwide": "worldwide_gross",
    "$Domestic": "domestic_gross",
    "$Foreign": "foreign_gross",
    "Domestic %": "domestic_pct",
    "Foreign %": "foreign_pct",
    "Genres": "genres",
    "Original_Language": "original_language",
    "Vote_Count": "vote_count",
    "Rating": "rating"
}, inplace=True)


df2.columns
```

--Clean Currency and Percentage Columns

```python

money_cols = ["worldwide_gross", "domestic_gross", "foreign_gross"]
for col in money_cols:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')

percent_cols = ["domestic_pct", "foreign_pct"]
for col in percent_cols:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')
```

```python
df2[col]
```

```python
--Standardize Text Columns
```

```python
for col in ["title", "genres", "original_language"]:
    df2[col] = df2[col].astype(str).str.strip().str.lower()
```

```python
df2.info()
```

```python
--Fix Rating Column
```

```python
import re
df2["rating"] = df2["rating"].astype(str).str.extract(r"(\d+\.\d+)").astype(float)
```

--Create a Normalized Title for Merging

```python
def normalize_title(title):
    if pd.isna(title):
        return None
    return re.sub(r'[^a-z0-9 ]', '', title.lower().strip())

df2["title_norm"] = df2["title"].apply(normalize_title)
```

```python
-- Keep Only Useful Columns
```

```python
useful_cols = [
    "title", "title_norm", "release_year",
    "worldwide_gross", "domestic_gross", "foreign_gross",
    "domestic_pct", "foreign_pct",
    "genres", "original_language",
    "vote_count", "rating"
]
df2 = df2[useful_cols]
```

```python
df2[useful_cols].columns
```

```python
df2.info()
df2.head()
```

```python
Merge Both Datasets
```

```python
merged_df = pd.merge(
    df_keep,
    df2,
    on=["title_norm", "release_year"],
    how="inner",  
    suffixes=('_movie', '_box')
)
merged_df 
```

```python
--Handle Duplicates or Missing Data
```

```python
merged_df.drop_duplicates(subset=['title_norm', 'release_year'], inplace=True)
merged_df.dropna(subset=['worldwide_gross', 'budget'], inplace=True)
```

```python
merged_df.to_csv("final_project_cleaned.csv", index=False)
print("✅ Cleaned and merged dataset saved successfully!")
```

--Find out Top 10 highest-grossing movies

```python
import pandas as pd
import matplotlib.pyplot as plt


merged_df = pd.read_csv('final_project_cleaned.csv')
```

```python
top10 = merged_df.sort_values(by='worldwide_gross', ascending=False).head(10)

top10.plot(
    x='title_movie', 
    y='worldwide_gross', 
    kind='barh', 
    figsize=(10,6), 
    color='Skyblue',
    title='Top 10 Highest Grossing Movies'
)
plt.xlabel('Worldwide Gross')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()   
plt.show()
```

-- Display Budget vs Worldwide Gross

```python
merged_df.plot(
    x='budget', 
    y='worldwide_gross', 
    kind='scatter', 
    figsize=(8,6), 
    color='teal',
    title='Budget vs Worldwide Gross'
)
plt.xlabel('Budget')
plt.ylabel('Worldwide Gross')
plt.show()
```

-- Display yearly Box Office Trend

```python
yearly = merged_df.groupby('release_year')['worldwide_gross'].sum()

yearly.plot(
    kind='line', 
    marker='o', 
    color='blue', 
    figsize=(10,5),
    title='Total Worldwide Gross by Year'
)
plt.xlabel('Release Year')
plt.ylabel('Total Worldwide Gross')
plt.show()
```

```python
--  find out the average Revenue by Genres_box
```

```python

merged_df['main_genre'] = merged_df['genres_box'].astype(str).str.split(',').str[0].str.strip()

genre_avg = merged_df.groupby('main_genre')['worldwide_gross'].mean().sort_values(ascending=False).head(10)

genre_avg.plot(
    kind='bar', 
    color='coral', 
    figsize=(10,6),
    title='Average Worldwide Gross by Genre'
)
plt.xlabel('Genre')
plt.ylabel('Average Worldwide Gross')
plt.show()
```

-- Display Rating vs Revenue Relationship

```python
merged_df.plot(
    x='rating', 
    y='worldwide_gross', 
    kind='scatter', 
    figsize=(8,6), 
    color='green',
    title='Audience Rating vs Worldwide Gross'
)
plt.xlabel('Rating')
plt.ylabel('Worldwide Gross')
plt.show()
```

-- find out top 10 most profitable movie

```python
merged_df['profit'] = merged_df['worldwide_gross'] - merged_df['budget']
top_profit = merged_df.sort_values(by='profit', ascending=False).head(10)

top_profit.plot(
    x='title_movie', y='profit', kind='barh', figsize=(10,6),
    color='mediumseagreen', title='Top 10 Most Profitable Movies'
)
plt.xlabel('Profit (Worldwide Gross - Budget)')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()
plt.show()
```

```python
-- find out Average Profit by Genre
```

```python
merged_df['main_genre'] = merged_df['genres_box'].astype(str).str.split(',').str[0].str.strip()

genre_profit = (
    merged_df.groupby('main_genre')['profit']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

genre_profit.plot(kind='pie', color='red', figsize=(10,6),
                  title='Average Profit by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Profit')
plt.show()
```

```python
-- find out total Worldwide Gross by Year
```

```python
yearly_gross = merged_df.groupby('release_year')['worldwide_gross'].sum()
yearly_gross.plot(kind='line', marker='o', figsize=(10,5), color='purple',
                  title='Total Worldwide Gross by Year')
plt.xlabel('Release Year')
plt.ylabel('Total Worldwide Gross')
plt.show()
```
